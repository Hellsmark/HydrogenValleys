# Packages and AI-function

```python
import os
from openai import OpenAI
import pandas as pd
from tqdm import tqdm
import ast

#openai.api_key = os.environ['OPENAI_API_KEY']
```

```python
# AI-function that cathegorises companies
client = OpenAI(api_key = 'API-KEY')
def cathegoriser(text):
    # Cathegories
    cath1 = 'Organisation type, which describes what kind of company it is, this can be one of the following: "Government" (including government agencies), "Municipality", "Firm" (includes both privately and state owned), "University" (and similar educational organisations), "Research institute" (non-unievrsities that perform research without providing education), "NGO". If none of these categories fit in any way you can categorise them as "NaN".'
    cath2 = 'Ownership type, which describes what kind of owner the company has, that is either publically owned by some state or municipality (that is tax-payed institution) or privately owned, as either "Private", "Public" or "Mix" (if there is both public and private owners of the firm).'
    cath3 = 'Industry sector, which describes what sector the company operates in, e.g. "energy", "chemicals", "manufacturing", "transport", etc.'
    cath4 = 'Nationality, which is from what country the company has its origin.'

    example = "[['Company1','Private company','Private','Energy provider (oil and gas)','Energy','Energy Provider,'Norway']['CompanyB','State owned','Public','Manufacturing (steel)','Process industry','Usage','Sweden']]"

    prompt = f"Classify the following companies: '{text}'"
    response = client.chat.completions.create(
        model="gpt-3.5-turbo-0125",
        messages=[{"role": "system", 
                   "content": "You are a helpful assistant that cathegorises companies depending on the following topics: 1."+cath1+' 2.'+cath2+' 3.'+cath3+' 4.'+cath4+'. You respond in the form of a python list where each company has its own list with the answer to the different topics as items in the list. Start each companylist with the name of the company. Here are is an example of how a answer can look like for two companies: '+example},
                  {"role":"user",
                   "content":prompt}],
        max_tokens=100,
        temperature=0.9
    )
    return response.choices[0].message.content.strip()

# . Please respond in the form of a python list where each company has its own list with the answer to the different topics as items in the list. Start each company list with the name of the company
```

# Load companies and call AI

```python
# Read a csv file where the companynames are present. In this case the top 10 recruiters. Then it calls for AI-function to cathegorise.
df = pd.read_csv('companies_for_AI_cathegorisation.csv')                    # -> Read a csv file of sort and make into a df
companies = list(df['Name'])

Result = []
print(type(df))
```

```python
for text in companies:
    print(text)
    answer = cathegoriser(text)
    Result.append(answer)
    print('ok')

print('ok')

df['Cathegories'] = Result
```

# Printing and saving as txt

```python
# This cell is for saving the result
f = open("new-company_cathegor_ai-0.9.txt", "w")
text = ''
for company in df['Cathegories']:
    splitted = company.split() 
    j = 0
    row = ''
    for c in splitted:
        if c == '1.' or c == '2.' or c == '3.' or c == '4.' or c == '5.':
            row += '\n'
            text += row
            row = c
        else:
            row += ' '+c
    text += row + '\n\n'
print(text)
f.write(text)
f.close()
```
