```python
import os
from openai import OpenAI
import pandas as pd
from tqdm import tqdm
```

```python
# AI-function that cathegorises companies
client = OpenAI(api_key = 'API-key')
def cathegoriser(text):
    prompt = f"Classify the following companies as where they are located in the valuechain (producer, storage, distribution and applier) of the hydrogen industry in the nordics: '{text}'"
    response = client.chat.completions.create(
        model="gpt-3.5-turbo-0125",
        messages=[{"role": "system", 
                   "content": "You are a helpful assistant that cathegorises companies depending on where they occur in the valuechain of hydrogen technology and hydrogen market"},
                  {"role":"user",
                   "content":prompt}],
        max_tokens=100,
        temperature=0.5
    )
    return response.choices[0].message.content.strip()
```

```python
# Read a csv file where the companynames are present. In this case the top 10 recruiters. Then it calls for AI-function to cathegorise.
df = pd.read_csv('Top_comp.csv')                    # -> Read a csv file of sort and make into a df
companies = list(df['Name'])

Positions = []

for text in companies:
    position = cathegoriser(text)
    Positions.append(position)

df['Value_chain_position'] = Positions
```

```python
# This cell is for printing the result
for company in df['Value_chain_position']:
    splitted = company.split() 
    j = 0
    for i in range(10,len(company),10):
        row = ''
        words = splitted[j:i]
        for word in words:
            row += ' '+word
        print(row)
        j = i   
```

```python
# This cell is for saving the result
f = open("company_cathegor_ai.txt", "a")
text = ''
for company in df['Value_chain_position']:
    splitted = company.split() 
    j = 0
    for i in range(10,len(company),10):
        row = ''
        words = splitted[j:i]
        for word in words:
            row += ' '+word
        row += '\n'
        text += row
        j = i   
    text += '\n'
f.write(text)
f.close()
```
